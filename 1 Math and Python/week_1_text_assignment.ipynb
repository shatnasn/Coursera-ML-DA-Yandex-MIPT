```python
import pandas as pd
import numpy as np
from scipy import spatial
import re
```

```python
with open('sentences.txt') as input_file:
    sentences = input_file.readlines()
```

```python
num_sentences = len(sentences)
```

```python
sup1 = ''
sup1 = sup1.join(sentences) #преобразование списка в строку
```

```python
sup1 = sup1.replace('\n', '') #удаляем переносы строк
print(sup1)
```

```python
sup1 = sup1.lower() #к нижнему регистру
print(sup1)
```

```python
import re
sup2 = re.split('[^a-z]', sup1) #разбиение
sup2 
```

```python
str_list = list(filter(None, sup2)) #удаление пустышек
print(str_list)
```

```python
words = {}
curr_index = 0
for sentence_token in str_list:
    for token in sentence_token:
        if token not in words:
            words[token] = curr_index
            curr_index += 1
```

```python
print(words, '\n')
print(str_list)
```

```python
num_words = len(words)
```

```python
matrix = np.zeros((num_sentences, num_words))
```

```python
for i in range(num_sentences):
    tokens = str_list[i]
    for token in tokens:
        matrix[i][words[token]] += 1
```

```python
distances = {}

first_sentence_metric = matrix[0, :]
for i in range(num_sentences):
    cmp_sentence_metric = matrix[i, :]
    
    distances[i] = spatial.distance.cosine(first_sentence_metric, cmp_sentence_metric)
```

```python
distances_df = pd.DataFrame.from_dict(distances, orient = 'index')
distances_df.columns = ['distance']
distances_df['sentence'] = (lambda x: sentences[x], distances_df.index.values)
```
